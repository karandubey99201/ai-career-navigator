```python
import pandas as pd
import json
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

print("Libraries loaded!")
print("Job Matching Engine is ready!")
```

```python
# Load job data and resume data
df_jobs = pd.read_csv(r'C:\Users\KARAN DUBEY\career_navigator\data\clean_jobs.csv')

# Load parsed resume data
with open(r'C:\Users\KARAN DUBEY\career_navigator\output\resume_parsed.json', 'r') as f:
    resume_data = json.load(f)

print(f"Jobs loaded: {len(df_jobs)}")
print(f"Resume skills: {resume_data['skills']}")
```

```python
# Combine resume skills into a single string
resume_skills_text = ' '.join(resume_data['skills'])
resume_full_text = resume_data['raw_text']

# Job descriptions list
job_descriptions = df_jobs['description'].fillna('').tolist()
job_titles = df_jobs['title'].tolist()
job_companies = df_jobs['company_name'].tolist()

print(f"Resume text ready: {len(resume_full_text)} characters")
print(f"Total jobs to match: {len(job_descriptions)}")
print("Starting job matching...")
```

```python
# Using TF-IDF Vectorizer for matching
print("AI matching in progress... please wait")

# Combine all documents - resume + all jobs
all_docs = [resume_full_text] + job_descriptions

# Build TF-IDF matrix
vectorizer = TfidfVectorizer(
    stop_words='english',
    max_features=5000,
    ngram_range=(1, 2)
)
tfidf_matrix = vectorizer.fit_transform(all_docs)

# Calculate cosine similarity
similarity_scores = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:])
scores = similarity_scores[0]

print(f"Matching complete!")
print(f"Highest match score: {max(scores)*100:.1f}%")
print(f"Average match score: {np.mean(scores)*100:.1f}%")
```

```python
# Find top 10 matching jobs
top_10_indices = np.argsort(scores)[::-1][:10]

print("="*60)
print("TOP 10 MATCHING JOBS FOR YOUR RESUME")
print("="*60)

top_jobs = []
for rank, idx in enumerate(top_10_indices, 1):
    job_info = {
        'rank': rank,
        'title': job_titles[idx],
        'company': str(job_companies[idx]),
        'match_score': round(scores[idx]*100, 1)
    }
    top_jobs.append(job_info)
    print(f"\nRank {rank}: {job_info['title']}")
    print(f"  Company: {job_info['company']}")
    print(f"  Match Score: {job_info['match_score']}%")

print("\n" + "="*60)
```

```python
# Save results
results_df = pd.DataFrame(top_jobs)
results_df.to_csv(
    r'C:\Users\KARAN DUBEY\career_navigator\output\top_matches.csv',
    index=False
)

print("Top matching jobs saved successfully!")
print(results_df)
```
