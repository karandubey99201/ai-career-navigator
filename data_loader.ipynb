```python
# Basic dataset info
print(f"Total rows: {len(df)}")
print(f"Total columns: {len(df.columns)}")
print(f"\nMissing values (important columns):")
print(df[['title', 'description', 'skills_desc', 'location']].isnull().sum())
```

```python
# Keep only important columns
df_clean = df[['job_id', 'title', 'company_name', 'location', 
               'description', 'skills_desc', 
               'formatted_experience_level', 'max_salary', 
               'min_salary', 'remote_allowed']].copy()

# Fill missing values
df_clean['skills_desc'] = df_clean['skills_desc'].fillna('Not specified')
df_clean['description'] = df_clean['description'].fillna('Not specified')
df_clean['location'] = df_clean['location'].fillna('Not specified')

# Save the file
df_clean.to_csv(r'C:\Users\KARAN DUBEY\career_navigator\data\clean_jobs.csv', index=False)

print(f"Clean data saved successfully!")
print(f"Total jobs: {len(df_clean)}")
print(f"\nSample data:")
print(df_clean.head(3))
```
